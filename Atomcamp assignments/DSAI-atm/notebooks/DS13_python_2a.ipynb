```python
# Program to demonstrate String slicing

#The [3:12] indicates that the string slicing will start from the 3rd index of the string to the 12th index,
#(12th character not including).

# Creating a String
String1 = "Learning Python for Data Science"

String1 = 'Creating a String'

print(String1[13])

print("Initial String: ")
print(String1)

# Printing 3rd to 12th character
print("\nSlicing characters from 3-12: ")

sub_string = String1[3:13]

print(String1[3:13])

# Printing characters between 3rd and 2nd last character

print("\nSlicing characters between 3rd and 2nd last character: ")
print(String1[3:-2])
```

**Working with Python String constants Built-In Functions**

String.lower() - Converts the entire string into lowercase letters.

String.upper() - Converts the entire string into uppercase letters.

String.endswith() - Returns True if a string ends with the given suffix otherwise returns False.

String.startswith() - Returns True if a string starts with the given prefix otherwise returns False.

String.count() - Returns the number of occurrences of a substring in the given string

String.isdigit() - Returns “True” if all characters in the string are digits, Otherwise, It returns “False”.

String.isalpha() - Returns “True” if all characters in the string are alphabets, Otherwise, It returns “False”.

String.index - Returns the position of the first occurrence of substring in a string.

String.swapcase() - Method converts all uppercase characters to lowercase and vice versa of the given string, and returns it.

String.replace() - Returns a copy of the string where all occurrences of a substring is replaced with another substring.

# **Adding elements**

```python
my_list = ["apple", "banana", "watermelon"]
print(my_list)
my_list.append("mango") #this operation directly changes the original list

print(my_list) #['apple', 'banana', 'watermelon', 'mango']

my_list.insert(2, "kiwi") #inserts an item at a specific index
print(my_list) #['apple', 'banana', 'kiwi', 'watermelon', 'mango']
my_list.insert(5, "pineapple")

print(my_list)

#append can only add ONE item. but extend can add multiple items.

#Note that extend takes in ONE argument, which is a list of items you wish to add

my_list.extend(["kiwi", "grape", "apricot"])
print(my_list)
```

**Removing elements**

```python
my_list=['apple', 'banana', 'kiwi', 'watermelon', 'mango', 'pineapple', 'kiwi', 'grape', 'mango','apricot']
print(my_list)
my_list.remove("mango")
print(my_list)

#this will remove the element at index 3 and return it to you. if left empty, it will remove the last element.
my_list.pop(3)
print(my_list)

#using pop(), you can even store the removed item in a new variable
removed_item = my_list.pop() #it will remove 1 value from the end of the list
print(my_list)
print(removed_item)
```

# Tuple
Tuples are used to store multiple items in a single variable.

Tuple is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Set, and Dictionary, all with different qualities and usage.

A tuple is a collection which is ordered and unchangeable.

Tuples are written with round brackets.

```python
thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

```python
thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
```

```python
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

```python
mytuple = ("apple", "banana", "cherry")
print(type(mytuple))
```

# There are four collection data types in the Python programming language:

List is a collection which is ordered and changeable. Allows duplicate members.

Tuple is a collection which is ordered and unchangeable. Allows duplicate members.

Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.

Dictionary is a collection which is ordered** and changeable. No duplicate members.
