```python
def calculate_product(*names):
    product = 1
    for num in names:
        product *= num
    return product

# Testing the function
print(calculate_product(2, 3, 4))  # Output: 24 (2 * 3 * 4)
print(calculate_product(5, 6))  # Output: 30 (5 * 6)
```

```python
def display_user_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Testing the function
display_user_info(name="John", age=25, city="New York")
# Output:
# name: John
# age: 25
# city: New York
```

```python
def test(**kwargs):
    for i in kwargs.values():
        print(i+2)

test(age=24,height=182)
```

```python
def summarize_info(*args, **kwargs):
    print("Information:")
    for item in args:
        print(f"Item: {item}")

    print("\nUser Details:")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Testing the function
summarize_info("Python", "Machine Learning", name="Alice", age=30, city="Boston")
```

**Python Anonymous/Lambda Function**

Introduce lambda functions as a shorthand for defining small, anonymous functions. They're especially useful for short operations that don't require a full function definition.

lambda: argument_list:expression


Lambda functions are primarily used for short, single-expression operations, particularly within higher-order functions like map, filter, and sorted. They are also useful for creating "throwaway" functions when a full function definition is not necessary.


Here's a more detailed explanation:

Anonymous functions:


Lambda functions are anonymous, meaning they don't have a name like regular functions defined with def.

Single expression:
They can only contain a single expression, which is evaluated and returned.

Higher-order functions:
They are often used with higher-order functions like map (to apply a function to each item in an iterable), filter (to select items based on a condition), and sorted (to sort items based on a specific key).

Readability:
Lambda functions can improve code readability when used appropriately for short, simple operations, keeping the code concise and focused.




Example:
x = lambda a : a + 10 defines a lambda function that adds 10 to its input

```python
# Example 1: Program for even numbers without lambda function

def even_numbers(nums):
    even_list = []
    for n in nums:
        if n % 2 == 0:
            even_list.append(n)
    return even_list

num_list = [10, 5, 12, 78, 6, 1, 7, 9]
ans = even_numbers(num_list)
print("Even numbers are:", ans)

```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Testing the function
print(factorial(5))  # Output: 120 (5 * 4 * 3 * 2 * 1)
print(factorial(3))  # Output: 6 (3 * 2 * 1)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):  # Check divisibility up to square root of num
        if num % i == 0:
            return False
    return True

# Testing the function
print(is_prime(11))  # Output: True
print(is_prime(10))  # Output: False
```

```python
def find_common_elements(list1, list2):
    common_elements = []
    for element in list1:
        if element in list2:
            common_elements.append(element)
    return common_elements

# Testing the function
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7]
print(find_common_elements(list1, list2))  # Output: [4, 5]
```
